```python
# імпорт бібліотек
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# загрузка зображення
img = cv2.imread('image.jpg')
```

```python
# візуілізація зображення 
plt.imshow(img)
plt.axis('off')
plt.show
```

### SOBEL

```python
# конвертація зображення в сірий спектр
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# настройка матриць фільтру Собеля
kernel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
kernel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
```

```python
# настройка фільтрів
edges_x = cv2.filter2D(gray, -1, kernel_x)
edges_y = cv2.filter2D(gray, -1, kernel_y)
edges = edges_x + edges_y
```

```python
# реалізація горизонтального та вертикального фільтрів
sobel_x = cv2.filter2D(gray, -1, kernel_x)
sobel_y = cv2.filter2D(gray, -1, kernel_y)
sobel = edges_x + edges_y
```

```python
# візуалізація фільтру Собеля в opencv
cv2.imshow('Original images', img)
cv2.imshow('SOBEL', sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# візуалізація фільтру Собеля в matplotlib
plt.imshow(sobel)
plt.axis('off')
plt.show
```

### CANNY

```python
# загрузка зображення
img = cv2.imread('image.jpg')
```

```python
# конвертація зображення в сірий спектр
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# настройка фільтра Canny
canny = cv2.Canny(gray, 50, 150)
```

```python
# візуалізація фільтру Canny в opencv
cv2.imshow('Original images', img)
cv2.imshow('CANNY', canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# візуалізація фільтру Canny в matplotlib
plt.imshow(canny)
plt.axis('off')
plt.show
```

### ROBERTS

```python
# загрузка зображення
img = cv2.imread('image.jpg')
```

```python
# конвертація зображення в сірий спектр
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# настройка матриць фільтру Робертса
roberts_kernel_v = np.array([[1, 0], [0, -1]], dtype=np.float32)  
roberts_kernel_h = np.array([[0, 1], [-1, 0]], dtype=np.float32)
```

```python
# реалізація горизонтального та вертикального фільтрів
roberts_x = cv2.filter2D(gray, -1, roberts_kernel_h)
roberts_y = cv2.filter2D(gray, -1, roberts_kernel_v)
roberts = roberts_x + roberts_y
```

```python
# візуалізація фільтру Робертса в opencv
cv2.imshow('Original images', img)
cv2.imshow('Roberts', roberts)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# візуалізація фільтру Робертса в matplotlib
plt.imshow(roberts)
plt.axis('off')
plt.show
```

```python
cv2.imshow('Original images', img)
cv2.imshow('SOBEL', sobel)
cv2.imshow('CANNY', canny)
cv2.imshow('Roberts', roberts)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
